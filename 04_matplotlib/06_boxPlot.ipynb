```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 폰트 설정
mpl.rc('font', family='Malgun Gothic')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
x= iris[iris.species == 'setosa']['sepal_length']
y= iris[iris.species == 'setosa']['sepal_width']
z= iris[iris.species == 'setosa']['petal_width']
w= iris[iris.species == 'setosa']['petal_width']
plt.boxplot([x,y,z,w], labels=['sepal_length','sepal_width','petal_width','petal_width'])

plt.show()
```
