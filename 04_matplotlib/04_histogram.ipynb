# 도수분포표(Histogram)
가로축은 계급, 세로축은 도수 (횟수나 개수)

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 폰트 설정
mpl.rc('font', family='Malgun Gothic')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight)
plt.show()
```

```python
np.random.seed(2020)
# np.random.randn 평균0, 표준편차 1의 정규분포를 따라 원소 1000개인 array
x = np.random.randn(1000)
plt.hist(x)
plt.show()
```

```python
np.random.seed(2020)
# np.random.randn 평균0, 표준편차 1의 정규분포를 따라 원소 1000개인 array
x = np.random.randn(100000)
plt.hist(x, bins=100) # 도수
plt.show()
```

```python
x = np.linspace(-4., 4., 800)
# 정규분포 식 만들기
nd = np.exp(-x*x/2) / np.sqrt(2*np.pi)
plt.plot(x, nd)
plt.axhline(0, alpha=0.5) ;plt.axvline(0, alpha=0.5)
plt.show()
```

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.legend()
plt.show()
```
