## 그래프 영역 채우기
- fill_between() : x축과 그래프 사이
- fill_betweenx() : y축과 그래프 사이
- fill() : 두 그래프 사이

- 투명도 alpha = 0 < < 1

```python
# fill_between(), fill_betweenx()

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label') ; plt.ylabel('Y-Label')

# plt.fill_between(x[1:3], y[1:3], alpha=0.5)                 ## fill_between() 사용
plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.8)  ## fill_betweenx() 사용

plt.axhline(0) ; plt.axvline(0)
# y=0 horizontal line ; x= 0 vertical line

plt.show()
```

- 수직선 그리기 (axhline/axvline)

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2.2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.axhline(1, 0, 0.55, color='gray', linestyle='--', linewidth='1')
plt.axvline(1, 0, 0.16, color='lightgray', linestyle=':', linewidth='2')

plt.axhline(5.83, 0, 0.95, color='gray', linestyle='--', linewidth='1')
plt.axvline(1.8, 0, 0.95, color='lightgray', linestyle=':', linewidth='2')

plt.show()
```

```python
# fill()

x = np.arange(1. , 4. ,0.01)
y1 = x ** 2
y2 = x ** 2 / 2

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

# x[100:300] 정의역 위에서 y2[100:300] - y1[100:200] 만큼의 치역 채움
# y2, y1의 치역은 같아야 함
plt.title('두 그래프 사이 채우기')
plt.fill_between(x[100:300], y1[100:300], y2[100:300], color='lightgray', alpha=0.5)

plt.show()
```

```python
# 임의 영역 채우기 

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

#  x, y 값의 리스트를 입력해주면, 각 x, y 점들로 정의되는 다각형(Polygon) 영역을 채운다
plt.title('임의 영역 채우기 ')
plt.fill([1.9, 1.9, 2.5, 3.1], [2, 5, 9.5, 8], color='lightgray', alpha=0.5)

plt.show()
```

## Sin, Cos 그래프 그리기

```python
from math import pi, sin, cos

font = { 'fontsize': 14 }

x = np.arange(0, 2* pi, 0.01)
sin = np.array(list(map(sin, x)))
cos = np.array(list(map(cos, x)))

plt.title('Sin함수와 Cos함수',fontdict=font)
plt.xlabel('Sin', fontdict=font) ; plt.ylabel('Cos' ,fontdict=font)

# Latex
plt.xticks(np.array([0, pi/2, pi, 3*pi/2, 2*pi]), labels=['0','π/2','1π','3π/2','2π'])

plt.yticks(np.array([-1,0,1]), labels=['Low bound','Zero','High bound'])

plt.axhline(1, color='gray', linestyle='--', linewidth='1')
plt.axhline(0, color='gray', linestyle='--', linewidth='1')
plt.axhline(-1, color='gray', linestyle='--', linewidth='1')

plt.axvline(pi/2, color='lightgray', linestyle=':', linewidth='2')
plt.axvline(pi, color='lightgray', linestyle=':', linewidth='2')
plt.axvline(3*pi/2, color='lightgray', linestyle=':', linewidth='2')


plt.plot(x, sin, label = 'sine')
plt.plot(x, cos, label = 'cosine')
plt.show()
```
