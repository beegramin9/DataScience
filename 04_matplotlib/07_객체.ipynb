### Axes 객체와 subplot 명령
ax인덱스 = subplot(전체 행 개수 , 전체 열 개수, 순서 (인덱스, 1부터 시작) ) <br>
총 그래프 개수 = 행의 개수 x 열의 개수 <br>



```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)

plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout() # 패딩 자동조절
plt.show()
```

### 한 번에 여러개 그리기 Subplots
ax = axes\[0, 1\]... 여기서 ax는 plt객체, 다시 말하면 plt.plot, plt.scatter 등등 모두 사용 가능

```python
fig, axes = plt.subplots(2,2)


np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

### 반복문으로 그리기


```python
np.random.seed(0)
fig, axes = plt.subplots(2,2, sharex=True, sharey = True)

# x,y축을 공유하기. 근데 나라면 안 할거 같아. 보기 힘들다

for i in range(2):
    for k in range(2):
        ax = axes[i][k]
        ax.plot(np.random.rand(5))
        axes[0, 0].set_title(f"axes {2*i+k}")

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0.0, 5.0, 100)

y1 = np.cos(2 * np.pi * x) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x)

fig, axes = plt.subplots(2,1, sharex=True)

# 2차원으로 갈 필요 없고 1차원

for i in range(2):
    ax = axes[i]
    ax.plot(x,y)
    y = y1 if i == 0 else y2
    plt.ylabel('Damped oscillation') if i == 0 else plt.ylabel('Undamped')
    plt.title('A tale of 2 subplots') if i == 0 else plt.title('time (s)')

plt.tight_layout() 
plt.show()
```

```python
# 위 그래프는 x,y 축 범위가 다르기 때문에 통일하기 위해서 set_limit을 해줘야 함

fig = plt.figure(figsize=(12,8))

nums = ['I','II','III','IV']

for i, num in enumerate(nums):
    ds = ans[ans.dataset == num]
    ax = plt.subplot(2,2,i+1)
    
    # 여기서는 왜 plot을 써야 scatter처럼 나올까? ==> 결국 위처럼 하나 아래처럼 하나 똑같이 matplotlib처럼 사용 가능
    # 그게 아니라 내가 plot에서 marker모양을 o으로 만들어줬으니까 그래

    ax.plot(ds['x'],ds['y'],'o')
    ax.set_title(f'Dataset {i+1}')
    ax.set_xlim((4,20)) ; ax.set_ylim( (3, 13) )
```

```python
fig, axes = plt.subplots(2,2,figsize=(12,10), sharex=True, sharey=True)
nums = ['I','II','III','IV']

for i, num in enumerate(nums):
    ds = ans[ans.dataset == num]
    x = ds['x'] ; y = ds['y']
    ax = axes[i//2][i%2] # i = 0,1,2,3 일때 인덱스가 [0][0], [0][1], [1][0], [1][1] 여야 함

    # 아니면 ax.scatter(x,y) 이렇게 써도 되고
    ax.plot(x,y,'o',markersize = 10)
    # regression
    ax.plot([3,20],[4.5,13], 'r-', lw = 2)
    ax.set_title(f'Dataset {num}')
    ax.set_ylim((3,13))
    ax.set_xlim((4,20))
```

### x축은 공유하지만, y축은 다른 경우 : twinx

```python
fig, ax0 = plt.subplots()
# x축은 공유하지만 y축은 다른, 다시 말해 한 x축에 두 개의 y축을 갖는 그래프 그리기
ax1 = ax0.twinx()

ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)

ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)

ax0.set_xlabel("공유되는 x축")
fig.legend(loc='upper right', bbox_to_anchor=(0.86, 0.89))
# 바깥에 생기는 legend 위치를 안으로 들어오도록 조절하기
plt.show()
```
